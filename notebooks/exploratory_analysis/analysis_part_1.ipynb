```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'start_date' to datetime format and extract year
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['year'] = df['start_date'].dt.year  # Extract year

# Group data by year and disaster subgroup, then count occurrences
disaster_counts = df.groupby(['year', 'disaster_subgroup']).size().reset_index(name='count')

# Plot the frequency of disasters over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=disaster_counts, x='year', y='count', hue='disaster_subgroup', marker='o')

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Number of Disasters")
plt.title("Frequency of Disasters Over the Years")
plt.legend(title="Disaster Subgroup")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is loaded into a DataFrame called df
# Replace 'country_column', 'year_column', and 'disaster_type_column' with actual column names from your dataset

# Convert the 'start_date' or equivalent column to datetime type (if it's not already)
df['start_date'] = pd.to_datetime(df['start_date'])

# Extract the year from 'start_date' column
df['year'] = df['start_date'].dt.year

# Group the data by country, year, and disaster type and count the occurrences
disaster_frequency = df.groupby(['iso_code', 'year', 'disaster_type']).size().reset_index(name='count')

# Visualizing the frequency of disasters over the years for each country
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.lineplot(data=disaster_frequency, x='year', y='count', hue='iso_code', markers=True)

plt.title('Frequency of Disasters Over Years by Country')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
