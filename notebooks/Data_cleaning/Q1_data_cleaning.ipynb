```python
# Drop irrelevant columns but keep 'Total Damage ('000 US$)'
dataset_1_cleaned = dataset_1.drop(columns=["Event Name", "Subregion", "End Month", "End Day", "Total Damage, Adjusted ('000 US$)"])

# Handle missing values by filling with appropriate values or dropping
dataset_1_cleaned['Total Deaths'] = dataset_1_cleaned['Total Deaths'].fillna(0)
dataset_1_cleaned['No. Injured'] = dataset_1_cleaned['No. Injured'].fillna(0)
dataset_1_cleaned['No. Affected'] = dataset_1_cleaned['No. Affected'].fillna(0)
dataset_1_cleaned['No. Homeless'] = dataset_1_cleaned['No. Homeless'].fillna(0)
dataset_1_cleaned['Total Affected'] = dataset_1_cleaned['Total Affected'].fillna(0)
dataset_1_cleaned['Insured Damage (\'000 US$)'] = dataset_1_cleaned['Insured Damage (\'000 US$)'].fillna(0)
dataset_1_cleaned['Total Damage (\'000 US$)'] = dataset_1_cleaned['Total Damage (\'000 US$)'].fillna(0)

# Remove rows with duplicates based on the 'DisNo.' column (assuming 'DisNo.' is a unique identifier)
dataset_1_cleaned = dataset_1_cleaned.drop_duplicates(subset=['DisNo.'])

# Standardize column names to lowercase and replace spaces with underscores
dataset_1_cleaned.columns = dataset_1_cleaned.columns.str.lower().str.replace(' ', '_')

# Ensure correct data types (convert numerical columns to floats or ints)
dataset_1_cleaned['total_deaths'] = dataset_1_cleaned['total_deaths'].astype(float)
dataset_1_cleaned['no._injured'] = dataset_1_cleaned['no._injured'].astype(float)
dataset_1_cleaned['no._affected'] = dataset_1_cleaned['no._affected'].astype(float)
dataset_1_cleaned['no._homeless'] = dataset_1_cleaned['no._homeless'].astype(float)
dataset_1_cleaned['total_affected'] = dataset_1_cleaned['total_affected'].astype(float)
dataset_1_cleaned['insured_damage_(\'000_us$)'] = dataset_1_cleaned['insured_damage_(\'000_us$)'].astype(float)
dataset_1_cleaned['total_damage_(\'000_us$)'] = dataset_1_cleaned['total_damage_(\'000_us$)'].astype(float)

# Preview the cleaned dataset
print("\nCleaned Dataset:")
print(dataset_1_cleaned.head())
# Save the cleaned dataset to the specified path
dataset_1_cleaned.to_csv(r'D:\work\college\E-Coders_SAS_Curiosity_Cup\datasets\Cleaned_data\cleaned_disaster_impact.csv', index=False)

print("Cleaned dataset saved successfully!")

```
